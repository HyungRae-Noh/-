## 데이터셋 개요

- **파일명**: `online_commerce_data.csv`
- **데이터 크기**: 147,311 행 × 14 열
- **형식**: CSV (콤마로 구분된 값)

## 칼럼 설명

| 칼럼명             | 데이터 타입 | 설명                                         |
|--------------------|-------------|----------------------------------------------|
| `customer_no`      | 정수        | 고객의 고유 번호                             |
| `age`              | 실수        | 고객의 나이                                  |
| `gender`           | 정수        | 고객의 성별 (1=남성, 2=여성, 9=미정)         |
| `order_no`         | 정수        | 주문의 고유 번호                             |
| `order_date`       | 문자열      | 주문이 이루어진 날짜                         |
| `channel`          | 문자열      | 상품 구매가 이루어진 채널 (예: TV, Mobile)   |
| `item_category`    | 문자열      | 구매된 상품의 카테고리 (예: 캐쥬얼, 레포츠패션)|
| `item_code`        | 정수        | 상품의 고유 코드                             |
| `item_name`        | 문자열      | 상품의 이름                                  |
| `price`            | 정수        | 상품의 단위 가격                             |
| `qty`              | 정수        | 구매된 상품의 수량                           |
| `order_amount`     | 정수        | 주문된 상품의 총 금액 (단가 × 수량)          |
| `discount_amount`  | 정수        | 주문에 적용된 할인 금액                      |
| `paid_amount`      | 정수        | 고객이 실제로 지불한 금액 (주문 금액 - 할인 금액) |
